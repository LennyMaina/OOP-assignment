Students class

```python
class Students:
    def __init__(self, name, course, marks):
        self.name = name
        self.course = course
        self.marks = marks

    def average_marks(self):
        if self.marks is None or len(self.marks) == 0:
            return 0
        average_marks = sum(self.marks) / len(self.marks)
        return average_marks

    def grade(self):
        grade = self.average_marks()
        if grade >= 70:
            return 'A'
        elif grade >= 60:
            return 'B'
        elif grade >= 50:
            return 'C'
        else:
            return 'D'
    
```

Product and store

```python
class Product:
    def __init__ (self, sku, name, category, price, quantity):
        self.sku = sku
        self.name = name
        self.category = category
        self.price = price
        self.quantity = quantity
    
    def total_value(self):
        return self.price * self.quantity

class Store:
    def __init__ (self):
        self.products = {}

    def add_product(self, product):
        if product.sku in self.products:
            return 'product exists'
        else: 
            self.products[product.sku] = product
            return 'product added'

    def sell(self, sku, qty):
        if sku not in self.products:
            return 'product not found'
        product = self.products[sku]
        if product.quantity < qty:
            return 'Insufficient product'
        else: 
            product.quantity -= qty
            return f'sold{qty} of {product.name}'

        def report(self):
            for product in self.products.values():
                print(f'{product.sku}, {product.name}, {product.qty}, {product.total_value}')
```

Sales report

```python
class SalesReport:
    def __init__(self, sku, product, quantity, name, price):
        self.sku = sku
        self.product = product
        self.quantity = quantity
        self.name = name
        self.price = price

    def total_value(self):
        return self.price * self.quantity


def revenue_by_product(transactions):
    totals = {}
    for t in transactions:  
        totals[t.sku] = totals.get(t.sku, 0) + t.total_value() 
    return totals


def top_n_products_by_revenue(transactions):
    totals = revenue_by_product(transactions)
    sorted_products = sorted(totals.items(), key=lambda x: x[1], reverse=True)
    return sorted_products
```
